```python
# create BFMatcher object
def match_descriptors(desc1, desc2):
    bf = cv2.BFMatcher()

    good_num = 0
    num_iter = 0
    for i in range(len(desc1)):
        for j in range(len(desc2)):
            num_iter+=1
            matches = bf.knnMatch(desc1[i],desc2[j], k=2)

            # Apply ratio test
            good = []
            for m,n in matches:
                if m.distance < 0.75*n.distance:
                    good.append([m])
            good_num+=(len(good)/len(matches))
    good_num/=num_iter
    return good_num
    
def find_max_match(descriptors_images):
    list_matches=[]
    for i,descriptor in enumerate(tqdm.tqdm(descriptors_images, position=0, leave=True)):
        max_descriptor = 0
        max_descriptor_match = 0
        for j,descriptor_second in enumerate(descriptors_images):
            print(j)
            if j==i:
                continue
            desc_matches = match_descriptors(descriptor, descriptor_second)
            if desc_matches>max_descriptor_match:
                max_descriptor_match = desc_matches
                max_descriptor = j
        list_matches.append((i,max_descriptor))
    return list_matches
list_matches = find_max_match(descriptors_images_exposures)
```
